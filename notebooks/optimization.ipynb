### Data Loading

```python
processor = FinanceDataProcessor()

bnd_data = processor.read_data(ticker='BND')
tsl_data = processor.read_data(ticker='TSLA')
sp_data = processor.read_data(ticker='SPY')
```

### Make forcasts

```python
data_list = [bnd_data, tsl_data, sp_data]
name = ['BND', 'TSLA', 'SPY']
```

```python
for index, data in enumerate(data_list):
    trainer = ForecastModel(data_frame=data)
    trainer.preprocess_data()
    trainer.train_auto_arima()
    trainer.train_seasonal_arima()
    trainer.train_lstm()
    trainer.make_prediction()
    trainer.save_predictions(file_path=f'../data/{name[index]}-predictions.csv')
```
